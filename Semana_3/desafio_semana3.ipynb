```python
df['estado_residencia'].unique().tolist()
```

```python
for uf in df['estado_residencia'].unique().tolist():
    plt.hist(df[df['estado_residencia'] == uf]['pontuacao_credito'],bins=20,rwidth=0.5)
    plt.show()
```

```python
dfPandas = df[['estado_residencia','pontuacao_credito']].sort_values(['estado_residencia','pontuacao_credito'],ascending=False).groupby('estado_residencia').pontuacao_credito.agg(media = 'mean', meadiana='median', desvio_padrao='std', moda=('pontuacao_credito', lambda x:x.value_counts().index[0]))
```

```python
dfPandas = dfPandas.T
```
